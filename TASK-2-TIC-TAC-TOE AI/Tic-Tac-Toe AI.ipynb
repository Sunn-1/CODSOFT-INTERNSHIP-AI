```python
import random

# Define the board as a 3x3 array
board = [['-', '-', '-'],
         ['-', '-', '-'],
         ['-', '-', '-']]

# Define the possible winning combinations as pairs of indices
winning_combinations = [
    [(0, 0), (0, 1), (0, 2)],
    [(1, 0), (1, 1), (1, 2)],
    [(2, 0), (2, 1), (2, 2)],
    [(0, 0), (1, 0), (2, 0)],
    [(0, 1), (1, 1), (2, 1)],
    [(0, 2), (1, 2), (2, 2)],
    [(0, 0), (1, 1), (2, 2)],
    [(0, 2), (1, 1), (2, 0)]
]

# Define the players
players = ['X', 'O']

# Define the function to print the board
def print_board(board):
    for row in board:
        print(' | '.join(row))
        print('-' * 9)

# Define the function to get the user's move
def get_user_move(board, current_player):
    while True:
        move = input("Player {} - Enter your move (1-9): ".format(current_player))
        if move.isdigit():
            move = int(move)
            if 1 <= move <= 9:
                row = (move - 1) // 3
                col = (move - 1) % 3
                if board[row][col] == '-':
                    return (row, col)
        print("Invalid move. Try again.")

# Define the function to check if there is a winner
def check_winner(board, current_player):
    for winning_combination in winning_combinations:
        if all(board[i][j] == current_player for i, j in winning_combination):
            return True
    return False

# Define the function to get the next player
def get_next_player(current_player):
    return 'X' if current_player == 'O' else 'O'

# Get the current player
current_player = random.choice(players)

# Start the game loop
while True:
    # Display the current player and the board to the user
    print("Current Player: Player {}".format(current_player))
    print_board(board)

    # Get the user's move
    move = get_user_move(board, current_player)

    # Update the board with the user's move
    board[move[0]][move[1]] = current_player

    # Check if the game is over
    if check_winner(board, current_player):
        print_board(board)
        print("Player {} wins!".format(current_player))
        break

    # Check for a tie (all cells are filled)
    if all(cell != '-' for row in board for cell in row):
        print_board(board)
        print("The game is a tie.")
        break

    # Switch to the next player
    current_player = 'X' if current_player == 'O' else 'O'
```
